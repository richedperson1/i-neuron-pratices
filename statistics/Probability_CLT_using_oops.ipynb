```python
import numpy as np
from pandas import read_csv
from sklearn.utils import resample
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import random 
np.random.seed(52)
```

```python
class Prob_stat:
    from sklearn.utils import resample

    def __init__(self, data, sample, unit):
        self.data = data
        self.sample = sample
        self.unit = unit
        self.std = np.std(data)

    def med(self):
        self.med1 = []
        for i in range(self.sample):
            sm = resample(self.data, n_samples=self.unit)
            self.med1.append(np.median(sm))
        print('---------------------Median value theorem---------------------')
        print("The sample median value is to be :----> ", np.median(self.med1))
        print('The population median is to be :------> ', np.median(self.data))
        low = np.percentile(self.med1, 2.5)
        high = np.percentile(self.med1, 97.5)
        print("For 95 percent C.I median is :-------->", f"[{low},{high}]")

    def clt(self, sample_num):
        self.mlt = []
        for i in range(sample_num):
            smv = resample(self.data, n_samples=self.unit)
            self.mlt.append(np.mean(smv))
        print('-------------------------Center Limit Theorem---------------------------')
        print("The sampling distribution mean is to be :----> ", np.mean(self.mlt))
        print("The population distribution mean is to be :--> ", np.mean(self.data))
        print("The 95% C.I for sampling mean is to be :----->",
              f"[{np.mean(self.mlt)-2*self.std} , {np.mean(self.mlt)+2*self.std}]")
        print('C.I by using Percentile')
        print("The 95% C.I for sampling mean is to be :----->",
              f"[{np.percentile(self.mlt,2.5)} , {np.percentile(self.mlt,97.5)}]")

    def plot(self, types):
        if int(types) == 1:
            self.clt(200)
            sns.histplot(self.mlt, bins=20, kde=True,
                         stat="density", common_norm=True)
            plt.show()
            sns.histplot(self.mlt, bins=20, kde=True, element="step",
                         stat="density", common_norm=True, cumulative=True, fill=True)
        else:
            self.med()
            sns.histplot(self.med1, bins=20, kde=True,
                         stat="density", common_norm=True)
            plt.show()
            sns.histplot(self.med1, bins=20, kde=True, element="step",
                         stat="density", common_norm=True, cumulative=True, fill=True)
```
