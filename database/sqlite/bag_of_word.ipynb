```python
from functools import reduce
import os
```

```python
file_name = os.listdir("./data_file")
```

```python
Input = "Wipro"
st = Input[:2]
n = len(Input)
a = st*n
print(a)
```

```python
try:
    dist = {}
    for file in file_name:
        try:
            with open(f"./data_file/{file}") as f:
                data = f.read()
                # print("open")
                for word in data.split():
                    if word in dist:
                        dist[word]+=1
                    else:
                        dist[word] = 1
            # return dist
        except Exception as e:
            print(e)
except Exception as e:
    print(e)
```
